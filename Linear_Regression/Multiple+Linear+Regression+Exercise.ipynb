## Create the regression

### Declare the dependent and the independent variables

```python
x1 = data[['size','year']]
y = data['price']
```

### Regression

```python
x = sm.add_constant(x1)
result = sm.OLS(y, x).fit()
result.summary()
```

This suggests that the both the independant variables equally effect the price 

#### To see the independant effect of each variable over the dependent variable we would see the graphical representation

```python
x2 = data['size']
xo = sm.add_constant(x2)
result2 = sm.OLS(y, xo).fit()
result2.summary()
```

```python
plt.scatter(x2, y)
regression_line = 1.019e+05 + 223.1787 * x2
plt.plot(x2, regression_line, color = 'red')
```

```python
x3 = data['year']
xo2 = sm.add_constant(x3)
result3 = sm.OLS(y, xo2).fit()
result3.summary()
```

```python
plt.scatter(x3, y)
regression_line = -2.758e+06 + 1515.5137 * x3
plt.plot(x3, regression_line, color = 'red')
```

#### Comparing the two models:

1. The model with both "size" and "year" has a higher adjusted R-squared value, indicating that it explains more of the variation in real estate prices compared to the model with only "size."

2. Including the "year" predictor seems to contribute to the model's explanatory power, suggesting that "year" provides additional information that helps improve the model's predictions.

### Dealing with the Dummy variable that is Attendance in this case which shows the attendance of student above or below 75 % with yes and no respectively

```python
data = pd.read_csv('C:/Users/HP/Downloads/Data_sets/1.03.+Dummies.csv')
data
```

```python
# WE mapped the yes with a new value that is 1 and no with a 0
data['Attendance'] = data['Attendance'].map({'Yes':1, 'No':0})
data.describe()
```

```python
data.describe()
```

```python
y = data['GPA']
x1 = data[['SAT', 'Attendance']]
x = sm.add_constant(x1)
result = sm.OLS(y, x).fit()
result.summary()
```
